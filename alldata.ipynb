```python
import matplotlib.pyplot as plt
import numpy as np

# Define the coordinates and their names
coords = [
    (0, 0), (0, 1), (2, 1), (0, 2), (0, 4),
    (1, 4), (2, 4), (0, 5), (1, 5), (2, 5),
    (0, 6), (1, 6), (0, 7), (0, 8), (1, 8)
]

labels = [
    'R1C0Q0', 'R1C1Q0', 'R1C1Q2', 'R1C2Q0', 'R1C4Q0', 'R1C4Q1', 'R1C4Q2',
    'R1C5Q0', 'R1C5Q1', 'R1C5Q2', 'R1C6Q0', 'R1C6Q1', 'R1C7Q0', 'R1C8Q0', 'R1C8Q1'
]

# Example sensor results to visualize with color mapping
sensor_results = np.random.rand(len(coords)) * 10  # Replace with actual values if available

# Extract x and y coordinates
x_vals = [point[0] for point in coords]
y_vals = [point[1] for point in coords]

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create the scatter plot with color mapping
sc = ax.scatter(x_vals, y_vals, c=sensor_results, cmap='coolwarm', s=200, edgecolors='black')

# Label each point
for (x, y), label in zip(coords, labels):
    ax.text(x, y, label, fontsize=9, ha='left', va='bottom')

# Add color bar
cbar = fig.colorbar(sc, ax=ax, orientation='vertical', shrink=0.8, pad=0.05)
cbar.set_label("Sensor Result Value")

# Label the axes
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')

# Add a title
ax.set_title('Scatter Plot of Sensor Locations with Color Mapping')

# Display the plot
plt.show()
```

```python
import numpy as np
from scipy.stats import pearsonr

# Convert lists to NumPy arrays if needed
x_array = np.array(x_vals)
y_array = np.array(y_vals)
results_array = np.array(sensor_results)

# Calculate Pearson correlation for x and sensor results
corr_x, p_val_x = pearsonr(x_array, results_array)
print(f"Correlation between X and results: {corr_x:.3f}, p-value: {p_val_x:.3f}")

# Calculate Pearson correlation for y and sensor results
corr_y, p_val_y = pearsonr(y_array, results_array)
print(f"Correlation between Y and results: {corr_y:.3f}, p-value: {p_val_y:.3f}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt

def plot_mini_plots_for_radius(radius_col="Dark current-r10.0"):
    """
    Loads all 15 device CSV files, extracts Voltage step and the specified radius column,
    then creates a single figure with mini-plots at each device's (x, y) position.
    
    Parameters
    ----------
    radius_col : str
        The column name for the radius data, e.g. "Dark current-r10.0", "Dark current-r20.0", etc.
    """
    # 1. Read all CSV files (adjust skiprows if needed)
    df1  = pd.read_csv('P22406121-3_R1C0Q0_VQCOP-B004_2025-02-16-17-06-57.csv',  skiprows=18)
    df2  = pd.read_csv('P22406121-3_R1C1Q0_VQCOP-B004_2025-02-16-17-13-28.csv',  skiprows=19)
    df3  = pd.read_csv('P22406121-3_R1C1Q2_VQCOP-B004_2025-02-16-17-02-07.csv',  skiprows=19)
    df4  = pd.read_csv('P22406121-3_R1C2Q0_VQCOP-B004_2025-02-16-17-17-40.csv',  skiprows=19)
    df5  = pd.read_csv('P22406121-3_R1C4Q0_VQCOP-B004_2025-02-16-16-42-43.csv',  skiprows=19)
    df6  = pd.read_csv('P22406121-3_R1C4Q1_VQCOP-B004_2025-02-16-16-48-20.csv',  skiprows=19)
    df7  = pd.read_csv('P22406121-3_R1C4Q2_VQCOP-B004_2025-02-16-16-57-42.csv',  skiprows=19)
    df8  = pd.read_csv('P22406121-3_R1C5Q0_VQCOP-B004_2025-02-16-16-32-20.csv',  skiprows=19)
    df9  = pd.read_csv('P22406121-3_R1C5Q1_VQCOP-B004_2025-02-16-16-37-27.csv',  skiprows=19)
    df10 = pd.read_csv('P22406121-3_R1C5Q2_VQCOP-B004_2025-02-16-16-52-53.csv',  skiprows=19)
    df11 = pd.read_csv('P22406121-3_R1C6Q0_VQCOP-B004_2025-02-16-16-20-04.csv',  skiprows=19)
    df12 = pd.read_csv('P22406121-3_R1C6Q1_VQCOP-B004_2025-02-16-16-25-37.csv',  skiprows=19)
    df13 = pd.read_csv('P22406121-3_R1C7Q0_VQCOP-B004_2025-02-16-16-15-25.csv',  skiprows=19)
    df14 = pd.read_csv('P22406121-3_R1C8Q0_VQCOP-B004_2025-02-16-16-00-16.csv',  skiprows=19)
    df15 = pd.read_csv('P22406121-3_R1C8Q1_VQCOP-B004_2025-02-16-16-04-57.csv',  skiprows=19)
    
    # 2. Create the devices array with (x, y) positions and relevant columns
    devices = [
        {
            "name": "R1C0Q0",
            "x": 0, 
            "y": 0,
            "voltage":          df1["Voltage step:"],
            "darkcurrent":      df1[radius_col],
        },
        {
            "name": "R1C1Q0",
            "x": 0,
            "y": 1,
            "voltage":          df2["Voltage step:"],
            "darkcurrent":      df2[radius_col],
        },
        {
            "name": "R1C1Q2",
            "x": 2,
            "y": 1,
            "voltage":          df3["Voltage step:"],
            "darkcurrent":      df3[radius_col],
        },
        {
            "name": "R1C2Q0",
            "x": 0,
            "y": 2,
            "voltage":          df4["Voltage step:"],
            "darkcurrent":      df4[radius_col],
        },
        {
            "name": "R1C4Q0",
            "x": 0,
            "y": 4,
            "voltage":          df5["Voltage step:"],
            "darkcurrent":      df5[radius_col],
        },
        {
            "name": "R1C4Q1",
            "x": 1,
            "y": 4,
            "voltage":          df6["Voltage step:"],
            "darkcurrent":      df6[radius_col],
        },
        {
            "name": "R1C4Q2",
            "x": 2,
            "y": 4,
            "voltage":          df7["Voltage step:"],
            "darkcurrent":      df7[radius_col],
        },
        {
            "name": "R1C5Q0",
            "x": 0,
            "y": 5,
            "voltage":          df8["Voltage step:"],
            "darkcurrent":      df8[radius_col],
        },
        {
            "name": "R1C5Q1",
            "x": 1,
            "y": 5,
            "voltage":          df9["Voltage step:"],
            "darkcurrent":      df9[radius_col],
        },
        {
            "name": "R1C5Q2",
            "x": 2,
            "y": 5,
            "voltage":          df10["Voltage step:"],
            "darkcurrent":      df10[radius_col],
        },
        {
            "name": "R1C6Q0",
            "x": 0,
            "y": 6,
            "voltage":          df11["Voltage step:"],
            "darkcurrent":      df11[radius_col],
        },
        {
            "name": "R1C6Q1",
            "x": 1,
            "y": 6,
            "voltage":          df12["Voltage step:"],
            "darkcurrent":      df12[radius_col],
        },
        {
            "name": "R1C7Q0",
            "x": 0,
            "y": 7,
            "voltage":          df13["Voltage step:"],
            "darkcurrent":      df13[radius_col],
        },
        {
            "name": "R1C8Q0",
            "x": 0,
            "y": 8,
            "voltage":          df14["Voltage step:"],
            "darkcurrent":      df14[radius_col],
        },
        {
            "name": "R1C8Q1",
            "x": 1,
            "y": 8,
            "voltage":          df15["Voltage step:"],
            "darkcurrent":      df15[radius_col],
        },
    ]

    # 3. Determine the wafer layout range
    all_x = [d["x"] for d in devices]
    all_y = [d["y"] for d in devices]
    x_min, x_max = min(all_x), max(all_x)
    y_min, y_max = min(all_y), max(all_y)

    # 4. Create the main figure/axes
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_xlim(x_min - 1, x_max + 1)
    ax.set_ylim(y_min - 1, y_max + 1)
    ax.set_xlabel("X Position")
    ax.set_ylabel("Y Position")
    ax.set_title(f"Mini-Plots of {radius_col} vs. Voltage at Each Device Position")
    ax.grid(True)

    # Optional: faint markers at device locations
    ax.scatter(all_x, all_y, s=100, color='lightgray')

    # 5. Create mini-plots (inset axes) for each device
    inset_width = 0.08  # tweak for bigger/smaller mini-plots
    inset_height = 0.08

    for d in devices:
        # Convert device (x, y) from data coords to figure coords
        device_display_coords = ax.transData.transform((d["x"], d["y"]))
        device_fig_coords = fig.transFigure.inverted().transform(device_display_coords)
        
        x_fig = device_fig_coords[0] - inset_width / 2
        y_fig = device_fig_coords[1] - inset_height / 2

        # Create an inset axis at that position
        ax_inset = fig.add_axes([x_fig, y_fig, inset_width, inset_height])

        # Plot the device's voltage vs. chosen radius
        ax_inset.plot(d["voltage"], d["darkcurrent"], marker='o', linestyle='-')
        ax_inset.set_title(d["name"], fontsize=8)
        ax_inset.tick_params(labelsize=6)

        # Optionally remove axis ticks if you want less clutter:
        # ax_inset.set_xticks([])
        # ax_inset.set_yticks([])

    plt.tight_layout()
    plt.show()

# ============ Usage Example ============
# Just call the function with the radius you want, e.g.:
# plot_mini_plots_for_radius("Dark current-r20.0")
```
